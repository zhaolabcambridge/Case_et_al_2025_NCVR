```python
matplotlib.rcParams.update({'font.size': 15})

# weblogos

fig = plt.figure(figsize=(20, 10))
gs = GridSpec(nrows=2, ncols=2, hspace=.15, top=1)

### exp

ax0 = fig.add_subplot(gs[0, 0])

subset = adata_sc
subset = subset[subset.obs["exp_stat"].isin(["expanded"])]

seqs = subset.obs.loc[:, ["IR_VDJ_1_junction_aa", "cc_aa_alignment_size"]]
seqs["len"] = None

for i, r, in seqs.iterrows():
    seqs.at[i, "len"] = len(r["IR_VDJ_1_junction_aa"])

print(seqs["len"].value_counts())

seqs["cut"] = None

seqs = seqs.loc[seqs["len"] >= 13]

for i, r, in seqs.iterrows():
    seqs.at[i, "cut"] = r["IR_VDJ_1_junction_aa"][4:12]
    
#seqs.to_csv("WL3_seqs/nonexpanded_seqs.csv")

m = motifs.create(seqs["cut"], "ARNDCQEGHILKMFPSTWYV")

logo_df = pd.DataFrame(m.counts)

# create Logo object

ss_logo = logomaker.Logo(logo_df,
                         width=.8,
                         vpad=.05,
                         ax=ax0,
                         #fade_probabilities=True,
                         stack_order='small_on_top',
                         font_name='Arial Rounded MT Bold',
                         color_scheme='NajafabadiEtAl2017')

ax0.set_title("Expanded CD4 Tregs")


ax1 = fig.add_subplot(gs[0, 1])


subset = adata_sc
subset = subset[subset.obs["exp_stat"].isin(["nonexpanded"])]

seqs = subset.obs.loc[:, ["IR_VDJ_1_junction_aa", "cc_aa_alignment_size"]]
seqs["len"] = None

for i, r, in seqs.iterrows():
    seqs.at[i, "len"] = len(r["IR_VDJ_1_junction_aa"])

print(seqs["len"].value_counts())

seqs["cut"] = None

seqs = seqs.loc[seqs["len"] >= 13]

for i, r, in seqs.iterrows():
    seqs.at[i, "cut"] = r["IR_VDJ_1_junction_aa"][4:12]

m = motifs.create(seqs["cut"], "ARNDCQEGHILKMFPSTWYV")

logo_df = pd.DataFrame(m.counts)

# create Logo object

ss_logo = logomaker.Logo(logo_df,
                         width=.8,
                         vpad=.05,
                         ax=ax1,
                         #fade_probabilities=True,
                         stack_order='small_on_top',
                         font_name='Arial Rounded MT Bold',
                         color_scheme='NajafabadiEtAl2017')

ax1.set_title("Nonexpanded CD4 Tregs")
```
